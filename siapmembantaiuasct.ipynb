#### Pembahasan
Komputer tidak bisa memahami perintah secara cepat seperti manusia, komputer perlu mengeksekusi perintah tiap baris satu per satu dimulai dari urutan teratas sampai terbawah.

Dalam kasus soal diatas menggunakan **conditional statement** yang terdiri dari `if`, `else if`, `else`

- `if` digunakan untuk mengecek kondisi apakah benar atau salah
- `else if` digunakan untuk mengecek kondisi lain jika kondisi sebelumnya belum sesuai
- `else` digunakan untuk mengeksekusi perintah dengan kondisi selain if dan else if, biasanya else berada di paling bawah

Hasil dari conditional statement adalah True atau False, jika True maka perintah atau kode di dalamnya akan tereksekusi. Selain itu jika salah maka komputer akan mengecek baris kondisional di bawahnya

Balik ke soalnya, dari soal di atas kita hanya perlu mencatat dua digit terakhir dari NIM masing-masing orang
1. Adi = 34
2. Budi = 44
3. Angel = 19
4. Manda = 33
5. Toni = 84

**Perlu diingat !** 
- Modulo atau mod adalah sisa dari pembagian. Misal `5 mod 3 = 2` dan `2 mod 3 = 2` (a mod b, jika a<b hasilnya adalah a)
- Modulo atau mod biasanya diberi dengan simbol `%`
- Tanda `==` ini untuk membandingkan suatu nilai apakah True (Benar) atau False (Salah). Bedakan dengan tanda `=`, tanda `=` dalam pemrograman berfungsi untuk memberikan suatu nilai terhadap variabel.


| x   | if (x % 4 == 1) | else if (x % 4 == 2) | else if (x % 4 == 3) | else (if x >= 50) | else     |
|-----|-----------------|----------------------|----------------------|-------------------|----------|
| 34  | False           | `True`               | -                    | -                 | -        |
| 44  | False           | False                | False                | False             | `True`   |
| 19  | False           | False                | `True`               | -                 | -        |
| 33  | `True`          | -                    | -                    | -                 | -        |
| 84  | False           | False                | False                | `True`            | -        |

Maka urutan yang akan tercetak = **Exercise 2, Exercise 5, Exercise 3, Exercise 1, Exercise 4**

> **Jawaban: E**

````python
# Cek Jawaban dalam format pemrograman Python (.py)
nim = 34, 44, 19, 33, 84
for x in nim:
    if x % 4 == 1:
        print("Exercise 1")
    elif x % 4 == 2:  # `elif` ini sama dengan else if, tapi kalau di soal uas nanti formatnya tetap dalam `else if` yah
        print("Exercise 2")
    elif x % 4 == 3:
        print("Exercise 3")
    else:
        if x >= 50:
            print("Exercise 4")
        else:
            print("Exercise 5")
````

4. karena syarat ketiga, maka  `while (i > 0)`, tujuannya agar i berhenti di angka 1 (jika `while (i >= 0)`, pengulangan i=0 saat dikurangi 1, hasilnya tidak akan mendapatkan 0)

```
i = 5
while (i > 0)
{
    i=i-1;
    write(i);
};
```

Bentuk soal di atas adalah *functional programming* yang menerapkan rekursi. Rekursi adalah proses suatu fungsi memanggil dirinya sendiri. 

Terdapat *conditional statement* yang dapat dilihat dari baris `if(a<c and b<c)` dan baris `else`.

Baris `misteri(30,20,10)` artinya akan memanggil fungsi `misteri(a,b,c)`, yang mana `a=30, b=20, c=10`.

Perhatikan baris `if(a<c and b<c)` artinya agar perintah dibawahnya dapat tereksekusi, maka kedua syaratnya yaitu `a<c` dan `b<c` harus sesuai dan bernilai TRUE barulah bagian perintah `if` akan dijalankan. 

Selama nilai `a>c` atau `b>c`, fungsi di atas akan selalu menjalankan baris `else`. Setiap baris `else` dijalankan, fungsi misteri akan dipanggil dengan syarat a dikurangi 1, b dikurangi 2, dan c ditambah dengan 3.

|fungsi|a|b|c|if(a<c and b<c)|else|
|:---:|:---:|:---:|:---:|:---:|:---:|

|misteri(29,18,13)|29|18|13|FALSE|47|
|misteri(28,16,16)|28|16|16|FALSE|44|
|misteri(27,14,19)|27|14|19|FALSE|41|
|misteri(26,12,22)|26|12|22|FALSE|38|
|misteri(25,10,25)|25|10|25|FALSE|35|

**Perlu Diingat!**

Bisa saja dalam bentuk soal lain notasi `AND` dan `OR` yang digunakan berbeda
- Notasi atau simbol lain dari `and` adalah `&&`
- Notasi atau simbol lain dari `or` adalah `||`
- Notasi atau simbol lain dari `not` adalah `!`
- Sama seperti MBL, `and` akan bernilai TRUE jika kedua nilai bernilai benar
- Sama seperti MBL, `or` akan bernilai TRUE jika salah satu nilainya bernilai benar
- Sama seperti MBL, `not` adalah negasi

````python
def misteri(a, b, c):
    if a<c and b<c:
        return c
    else:
        return a + b + misteri(a-1, b-2, c+3)

x = misteri(30, 20, 10)
print(x)
````
